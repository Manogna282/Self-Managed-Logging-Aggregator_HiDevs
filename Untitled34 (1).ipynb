```python
import asyncio, aiosqlite, aiohttp, uvloop
import os, re, json, time, gzip, sqlite3
from datetime import datetime
import pandas as pd
import numpy as np
from sklearn.ensemble import IsolationForest
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# runtime config for GitHub Codespaces
BASE_DIR = "/workspaces/Self-Managed-Logging-Aggregator_HiDevs"

DB_PATH = f"{BASE_DIR}/logs.db"
LOG_FOLDER = f"{BASE_DIR}/logs"
os.makedirs(LOG_FOLDER, exist_ok=True)

# simple regex for parsing: timestamp level component message
LOG_REGEX = re.compile(r"^(?P<timestamp>[^ ]+)\s+(?P<level>INFO|WARN|ERROR|DEBUG)\s+(?P<component>[^:]+):\s+(?P<message>.*)$")
```

```python
def init_db(db_path=DB_PATH):
    con = sqlite3.connect(db_path)
    cur = con.cursor()
    cur.execute('''
    CREATE TABLE IF NOT EXISTS logs (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    ts TEXT,
    ts_unix REAL,
    level TEXT,
    component TEXT,
    message TEXT,
    raw TEXT
    )
    ''')
    cur.execute('''
    CREATE TABLE IF NOT EXISTS metrics (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    name TEXT,
    value REAL,
    ts TEXT
    )
    ''')
    con.commit()
    con.close()


init_db()
print('DB initialized at', DB_PATH)
```

```python
import random
from datetime import datetime, timezone

LEVELS = ['INFO','WARN','ERROR','DEBUG']
COMPONENTS = ['auth','api','db','worker','scheduler']

async def generate_logs(path, rate_per_sec=5, runtime_sec=30):
    path = os.path.join(LOG_FOLDER, path)
    end = time.time() + runtime_sec
    with open(path, 'a') as f:
        while time.time() < end:
            ts = datetime.now(timezone.utc).isoformat().replace('+00:00', 'Z')

            lvl = random.choices(LEVELS, weights=[0.6,0.2,0.1,0.1])[0]
            comp = random.choice(COMPONENTS)
            msg = f"{comp} processed request id={random.randint(1,10000)} status={'ok' if lvl!='ERROR' else 'failed'}"
            line = f"{ts} {lvl} {comp}: {msg}\n"
            f.write(line)
            f.flush()
            await asyncio.sleep(1.0 / rate_per_sec)
    return path
```

```python
async def tail_file(path, queue, poll_interval=0.2):
    # open and seek to end
    with open(path, 'r') as f:
        f.seek(0, os.SEEK_END)
        while True:
            line = f.readline()
            if not line:
                await asyncio.sleep(poll_interval)
                continue
            await queue.put((line, time.time()))
```

```python
import socket


async def udp_syslog_listener(host='0.0.0.0', port=1514, queue=None):
    loop = asyncio.get_event_loop()
    sock = socket.socket(socket.AF_INET, socket.SOCK_DGRAM)
    sock.bind((host, port))
    sock.setblocking(False)
    print('UDP syslog listening on', host, port)
    while True:
        data, addr = await loop.run_in_executor(None, sock.recvfrom, 65535)
        line = data.decode('utf-8', errors='ignore').strip()
        await queue.put((line, time.time()))


from aiohttp import web


async def http_ingest(app):
    async def handle(request):
        payload = await request.text()
        await app['queue'].put((payload, time.time()))
        return web.Response(text='ok')
    return handle
```

```python
async def parse_line(line):
    m = LOG_REGEX.match(line.strip())
    if m:
        ts = m.group('timestamp')
        try:
            ts_unix = datetime.fromisoformat(ts.replace('Z','')).timestamp()
        except Exception:
            ts_unix = time.time()
        return {
            'ts': ts,
            'ts_unix': ts_unix,
            'level': m.group('level'),
            'component': m.group('component'),
            'message': m.group('message'),
            'raw': line.strip()
        }
    else:
        # fallback: store raw
        return {
            'ts': datetime.now(datetime.UTC).isoformat().replace('+00:00', 'Z'),
            'ts_unix': time.time(),
            'level': 'UNKNOWN',
            'component': 'unknown',
            'message': line.strip(),
            'raw': line.strip()
        }


async def db_writer_worker(queue, db_path=DB_PATH):
    # single DB writer to avoid concurrency issues
    con = await aiosqlite.connect(db_path)
    while True:
        item = await queue.get()
        if item is None:
            break
        parsed = await parse_line(item[0])
        # insert
        await con.execute('INSERT INTO logs (ts, ts_unix, level, component, message, raw) VALUES (?,?,?,?,?,?)',
                          (parsed['ts'], parsed['ts_unix'], parsed['level'], parsed['component'], parsed['message'], parsed['raw']))
        await con.commit()
        queue.task_done()
    await con.close()
```

```python
async def run_demo():
    q = asyncio.Queue()
    # start DB writer
    writer_task = asyncio.create_task(db_writer_worker(q))
    # start log generator
    gen_task = asyncio.create_task(generate_logs('demo.log', rate_per_sec=8, runtime_sec=20))
    # start tail
    tail_task = asyncio.create_task(tail_file(os.path.join(LOG_FOLDER,'demo.log'), q))
    await asyncio.gather(gen_task)
    # wait a few seconds to flush
    await asyncio.sleep(2)
    # stop tail & writer
    tail_task.cancel()
    await q.put(None)
    await writer_task
```

```python
import sqlite3

def load_logs(limit=5000):
    con = sqlite3.connect(DB_PATH)
    df = pd.read_sql_query('SELECT id, ts, ts_unix, level, component, message FROM logs ORDER BY ts_unix DESC LIMIT ?', con, params=(limit,))
    con.close()
    return df


# text vectorization

def analyze_logs():
    df = load_logs(2000)
    if df.empty:
        print('no logs')
        return
    vec = TfidfVectorizer(max_features=1000, stop_words='english')
    X = vec.fit_transform(df['message'].astype(str))


    # clustering for pattern discovery
    k = min(8, max(2, int(len(df)**0.5)))
    km = KMeans(n_clusters=k, random_state=42).fit(X)
    df['cluster'] = km.labels_


    # anomaly detection on vector features
    iso = IsolationForest(contamination=0.02, random_state=42)
    iso.fit(X.toarray())
    df['anomaly_score'] = iso.decision_function(X.toarray())
    df['is_anomaly'] = iso.predict(X.toarray()) == -1


    print('Clusters:', df['cluster'].value_counts().to_dict())
    print('Anomalies:', df['is_anomaly'].sum())
    return df


# Usage: df = analyze_logs()
```

```python
def compute_latency_samples(sample_count=1000):
    con = sqlite3.connect(DB_PATH)
    cur = con.cursor()
    cur.execute('SELECT ts_unix FROM logs ORDER BY ts_unix DESC LIMIT ?', (sample_count,))
    rows = cur.fetchall()
    con.close()
    if not rows:
        return None
    now = time.time()
    latencies = [now - r[0] for r in rows]
    return np.array(latencies)


lat = compute_latency_samples(500)
if lat is not None:
    print('latency p50,p90,p99:', np.percentile(lat,[50,90,99]))
    plt.figure(figsize=(6,3))
    plt.hist(lat, bins=50)
    plt.title('Log ingestion latency (s)')
    plt.xlabel('seconds')
    plt.show()
```
